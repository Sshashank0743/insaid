### 5.1 Which are the top 10 brands on sale in 1953-2016 ?

```python
sns.countplot(x='Brand', data=cardata, order=cardata['Brand'].value_counts().nlargest(10).index)
plt.title('Top 10 Brands by sale volume', fontsize=16)
plt.xlabel("Brand Name", fontsize=16)
plt.ylabel('No. of vehicles', fontsize=16)
plt.xticks(rotation=60)
plt.savefig('images/Top 10 Brands by sale volume.png')
```

```python
cardata.Brand.value_counts().nlargest(10).plot.pie(autopct='%.2f', figsize=(6,6))
plt.ylabel('')
plt.title('Top 10 Brands by sale volume %', fontsize=16)
plt.savefig('images/Top 10 Brands by sale volume percent.png')
```

This shows __Volkswagen and Mercedes-Benz__ are __top__ brands on sale and hence these would be preferred choices for people

<a id=section502></a>
### 5.2 Which are the most preferred body type used in 1953-2016 ?

```python
sns.countplot(x='body', data=cardata, order=cardata['body'].value_counts().index)
plt.title("Vehicles sold by Body Type", fontsize=16)
plt.xlabel("Body Type", fontsize=16)
plt.ylabel("No. Of Vehicles", fontsize=16)
plt.savefig('images/Vehicles sold by Body Type.png')
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(6,6)
cardata['body'].value_counts().plot(kind='pie', autopct='%.2f')
plt.ylabel('')
plt.title("% Vehicles sold by Body Type", fontsize=16)
plt.savefig('images/Vehicles sold by Body Type percent.png')
```

This chart says that **"Sedan"** body type cars are the most sold cars, if we want to see it in terms of %, then we can see it in above chart. We can say this in other terms that the most preferred choice for people is Sedan, because it is the perfect and comfortable car for small size family within lower budget.

### 5.6 Co-relation between Brand and Body type in lower price segment.

```python
bins = [0, 20000,40000,60000,80000,100000, 600000]
lab = ['<20K','20-39K','40-59K','60-79K','80-99K','>100K']
cardata['price_group'] = pd.cut(cardata.price,bins=bins, labels=lab, include_lowest=True)
cardata.price_group = cardata.price_group.astype(object)
cardata.head()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15, 3)

lowvalue = cardata[cardata.price_group == '<20K']
lst = lowvalue.Brand.value_counts().nlargest(15).index
sns.countplot(x='Brand', data=lowvalue, order=lst, hue='body')

plt.xticks(rotation=50)
plt.title("Brand and Body type co-relation", fontsize=16)
plt.xlabel("Car Brand", fontsize=16)
plt.ylabel("No. of Vehicles", fontsize=16)
plt.legend(bbox_to_anchor=(0.9,0.45))

plt.savefig('images/brand_bodytype_corelation_low.png')
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(6,6)
lowvalue.Brand.value_counts().nlargest(10).plot(kind='pie', autopct="%.2f")
plt.title("Brand wise sale distribution %", fontsize=16)
plt.ylabel('')

plt.savefig('images/Brand wise sale distribution percent.png')
```

From above chart, it is observed that **sedan** type of cars are the highest selling cars in lower price segment for all the brands except for **Volkswagon** and **Renault**, which have **Van** type of cars selling more.

In lower price segment leading brands for each body type are as below :<br>
- *Sedan* : **BMW**, closely followed by VAZ and Mercedes-Benz<br>
- *Van* :  **Volkswagon**, followed by Renault and Mercedes-Benz<br>
- *Vagon* : **Renault**, followed by Audi and Volkswagon<br>
- *Crossover* : **Mitsubishi and Nissan** followed by Hyundai<br>
- *Hatch* : **Ford** followed by VAZ and Volkswagon<br>

### 5.14 Distribution of No of cars sold and the Total value of them per annum

```python
cars = cardata[cardata['prod_year']>2000].groupby('prod_year').agg({'price':['sum', 'count']}).reset_index()

cars.columns = ['year', 'Total_Amount', 'Number_of_Vehicles']
cars['stdamt'] = cars['Total_Amount']/10000

fig, ax = plt.subplots()
fig.set_size_inches(10,5)

line1, = plt.plot(cars.year,cars.stdamt, 'r', label = 'line 1')
line2, = plt.plot(cars.year,cars.Number_of_Vehicles, 'b', label = 'line 2')

plt.legend([line1, line2],['Standardised Total Value of Cars Sold','Total Number of Cars Sold'])
plt.title('Distribution of No of cars sold & the Total value of per annum', fontsize=16)
plt.ylabel('Average Price', fontsize=16)
plt.xlabel('Production Year', fontsize=16)

plt.savefig('images/Distribution of No of cars sold.png')
```

People are preferring high cost luxury cars from 2015 as seen from the quite drastic jump in the total value of cars sold between 2015 and 2016.

### 5.32 When is body most likely to be "hatch" type ? (body vs brand)

```python
df = pd.DataFrame(columns = ['Brand','avg','percent'])

bodyType='hatch'

avg = cardata[cardata.body == bodyType].Brand.value_counts()/len(cardata)*100
avg = avg.sum()
```

```python
brandValue='VAZ'

VAZ_crossover = cardata[(cardata.body == bodyType) & (cardata.Brand == brandValue)].Brand.value_counts()
VAZ = cardata[(cardata.Brand == brandValue)].Brand.value_counts()
perct = VAZ_crossover/VAZ*100

df = df.append({
     "Brand":brandValue,
     "avg": avg,
     "percent":  perct[0]
      }, ignore_index=True)
```

```python
filteredvalue = cardata[(cardata.body == bodyType) & ~(cardata.Brand == brandValue)].Brand.value_counts()
actual = cardata[~(cardata.Brand == brandValue)].Brand.value_counts()
perct = filteredvalue.sum()/actual.sum()*100
    
df = df.append({
 "Brand":"Not VAZ",
 "avg": avg,
 "percent":  perct
  }, ignore_index=True)
```

- When Brand is __"VAZ", 22.74%__ of body is __"hatch"__, this is __~10% higher__ than the average sale percentage (13.21%).

```python
def get_avg_brand_pecent(bodyType, brandValue):
    avg_df = cardata[cardata.body == bodyType].Brand.value_counts()/len(cardata)*100
    avg = avg_df.sum()
    
    filteredvalue = cardata[(cardata.body == bodyType) & (cardata.Brand == brandValue)].Brand.value_counts()
    actual = cardata[(cardata.Brand == brandValue)].Brand.value_counts()
    perct = filteredvalue/actual*100
    return avg, perct
```

### 6. Conclusion <a id=section6></a>

- In this dataset We have total 9463 observations after removing duplicate records.
- We observe that out of total cars sold, majority of the cars sold by **Volkswagen, Mercedes-Benz and BMW** brand.
- **Sedan, Crossover and Hatch** body types cars are the highest sold cars.
- **Front wheel drive** cars are the highest sold cars, because most of the cars of this segment falls in lower price range.
- **Front wheel drive** cars are the top selling cars in all body type except for **Crossover** body type. In **Crossover** body type **Full wheel drive** cars are leading.
- **80%** of total cars sold in the price range of **200 to 20000** with body type as **Sedan**, drive type as **Front** and engine type as **Petrol**, because the middle class families preferred low price cars with Sedan body and Petrol engine type (ease of maintainence).
- **Mercedes-Benz** is the leading brand in higher price segment cars.
- Majority of the **higher price range** cars sold with body type as **Crossover**, drive type as **Full** and engine type as **Diesel**.
- All body segment of cars have **more average price for Full drive** type cars compare to other drive type cars.
- All engine types of cars have **more average price for Full drive** type cars compare to other drive type cars.
- Approx **55%** of total cars sold have engine volume in the range of **1.5 CC to 2.5 CC**.
- Most of the cars sold in **2.0 CC** engine volume segment followed by **1.5 CC**, there are **very few cars sold below 1.0 CC and above 4.0 CC**.
- It is observed that **lower price segment car has higher mileage** and higher price segment car has lower mileage.
- Majority of the cars with **production year 2008 or later were registered** and sold at **price of 50000 or above**.
- Production year **2008 and 2012** cars are the highest sold cars.
- **Price** has **positive correlation** with **engine volume and production year** and **negative** correlation with **mileage**.
- **Production year** has **negative correlation** with **mileage and engine volume**.
